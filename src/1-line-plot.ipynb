### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.arange(0, 11, 1)
# Generate y values as the square of x
y = x ** 2

# Create the plot
plt.plot(x, y)

# Add title and labels
plt.title('Plot of y = x^2')
plt.xlabel('x values')
plt.ylabel('y values')

# Add grid
plt.grid(True)

# Display the plot
```
